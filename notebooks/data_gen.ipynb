```python
import cv2
from mtcnn import MTCNN

font = cv2.FONT_HERSHEY_SIMPLEX

def find_face_MTCNN(frame, result_list):
    for result in result_list:
        x, y, w, h = result['box']
        crop_img = frame[y:y+h, x:x+w]
        if len(result)>0:
            cv2.imwrite("../data/Shubhyansh/"+str(counter)+".jpg",frame)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,155,255) ,2)
    return frame


video_capture = cv2.VideoCapture(0)
detector = MTCNN()

counter = 1

while True:
    _, frame = video_capture.read()
    faces = detector.detect_faces(frame)
    detectFaceMTCNN = find_face_MTCNN(frame, faces)
    frame = cv2.putText(frame, str(counter), (50, 50), font, 1, (255, 255, 0), 2, cv2.LINE_AA)
    cv2.imshow('Video', detectFaceMTCNN)
    if len(faces)>0:
        counter = counter + 1
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    if counter > 30:
        break

video_capture.release()
cv2.destroyAllWindows()
```
